## Intracellular

```python
# Read the file
import numpy as np
import matplotlib.pyplot as plt

data = []
with open('CryptoRun_MOD_in.txt', 'r') as file:
    for line in file:
        entry = line.strip().split(',')
        data.append((int(entry[0]), float(entry[1])))

# Filter the data based on the conditions
y_data = []   # for time greater than 67
x_data = [] # for time =0
for entry in data:
    cn, time = entry
    if time >=64:
        y_data.append(cn)
    elif time == 0:
        x_data.append(cn)

#print (y_data)
#print(x_data)
# since they ave different size, we have to sort the data into diffren
min_len = min(len(x_data), len(y_data))
cn_time_67 = y_data[:min_len]
cn_time_0 = x_data[:min_len]

# Generate scatter plot for the picked data
plt.figure(figsize=(14,8))
fig, ax = plt.subplots(figsize = (9, 6))
ax.scatter(cn_time_0, cn_time_67, s=60, alpha=0.7, edgecolors="k")

# Set logarithmic scale on the both variables
ax.set_xscale("log")
ax.set_yscale("log");

plt.xlabel('Number of Cryptococcus Cells at 0hr')
plt.ylabel('Number of Cryptococcus Cells at 64hrs')
plt.title('Intracellular Population')

# Set custom x-axis and y-axis range and ticks
plt.xlim(1, 10**3+500)
plt.ylim(1, 10**5+500)
plt.savefig('two_Mod_intra.jpeg', dpi = 300)

plt.show()
```
