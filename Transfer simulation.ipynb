### Exponential test

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import expon
from scipy.stats import kstest

# Read data from 'data_1000_Notransfer.txt'
data = []
with open('data_1000_transfer.txt', 'r') as file:
    for line in file:
        entry = line.strip().split(',')
        if float(entry[1]) >= 64:
            data.append(int(entry[0]))
            
# Fit the data to an exponential distribution
lambda_fit = 1 / np.mean(data)

# Generate a random sample from the fitted exponential distribution
sample_fit = expon.rvs(scale=1/lambda_fit, size=len(data))

# Perform the KS test for goodness of fit
ks_stat, p_value = kstest(data, expon.cdf, args=(0, 1/lambda_fit))

# Print the KS test result
print("Exponential Distribution Test:")
print(f"KS statistic: {ks_stat}")
print(f"p-value: {p_value}")

# Check if the data follows an exponential distribution
if p_value > 0.05:
    print("The data follows an exponential distribution.")
else:
    print("The data does not follow an exponential distribution.")

    
plt.figure(figsize=(14, 8))
# Plot the data and fitted exponential distribution as a bar plot
bins = np.linspace(min(data), max(data), 20)
plt.hist(data, bins=bins, density=True, alpha=0.5, edgecolor='black', linewidth=1, label='Data')
plt.plot(np.sort(data), expon.pdf(np.sort(data), scale=1/lambda_fit), 'r', label='Exponential Distribution')
plt.xlabel('Number of Cn cells')
plt.ylabel('Probability Density')
plt.title('Exponential Distribution Fit')
plt.legend()
plt.savefig('Expo.jpeg', dpi=200)
plt.show()
```

## Normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import shapiro

# Read data from file
data = []
with open('data_1000_transfer.txt', 'r') as file:
    for line in file:
        entry = line.strip().split(',')
        if float(entry[1]) >= 64:
            data.append(int(entry[0]))

# Perform the Shapiro-Wilk test for normality
statistic, p_value = shapiro(data)
plt.figure(figsize=(14, 8))

# Plot the bar chart of the data
bin_edges = np.linspace(min(data), max(data), num=11)
bin_counts, _, _ = plt.hist(data, bins=bin_edges, density=True, alpha=0.5, edgecolor='black')
bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])
bin_width = bin_edges[1] - bin_edges[0]

# Add a normal distribution curve to the plot
mu, sigma = np.mean(data), np.std(data)
x = np.linspace(min(data), max(data), 100)
pdf = 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x - mu) / sigma)**2)

plt.plot(x, pdf, 'r-', label='Normal Distribution')

# Add labels and title to the plot
plt.xlabel('Number of Cryptococcus Cells')
plt.ylabel('Probability Density')
#plt.title('')

# Add a legend to the plot
plt.legend()

# Show the plot
plt.savefig('Normal.jpeg', dpi=200)
plt.show()

# Print the Shapiro-Wilk test result
print("Shapiro-Wilk Test:")
print(f"Test statistic: {statistic}")
print(f"p-value: {p_value}")

# Check if the data follows a normal distribution
if p_value > 0.05:
    print("The data follows a normal distribution.")
else:
    print("The data does not follow a normal distribution.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy.stats import norm, expon, gamma, lognorm
import numpy as np

# Read data from 'data_1000_Notransfer.txt'
data_1000_notransfer = []
with open('data_1000_Intra_Notransfer.txt', 'r') as file:
    for line in file:
        entry = line.strip().split(',')
        if float(entry[1]) >= 64:
            data_1000_notransfer.append(int(entry[0]))

# Read data from 'data_1000_Intra_transfer.txt'
data_1000_intra_transfer = []
with open('data_1000_Intra_transfer.txt', 'r') as file:
    for line in file:
        entry = line.strip().split(',')
        if float(entry[1]) >= 64:
            data_1000_intra_transfer.append(int(entry[0]))

# Set up the figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot KDE for transfer data
sns.kdeplot(ax=axes[0], data=data_1000_intra_transfer, color='blue', fill=True)
axes[0].set_xlabel('Number of Cryptococcus Cells')
axes[0].set_ylabel('Density')
axes[0].set_title('Transfer Data Distribution (KDE)')

# Plot KDE for no transfer data
sns.kdeplot(ax=axes[1], data=data_1000_notransfer, color='green', fill=True)
axes[1].set_xlabel('Number of Cryptococcus Cells')
axes[1].set_ylabel('Density')
axes[1].set_title('No Transfer Data Distribution (KDE)')

# Fit KDE to known distribution models and select the best fit
kde_transfer = sm.nonparametric.KDEUnivariate(data_1000_intra_transfer)
kde_transfer.fit()
mean_transfer = np.mean(data_1000_intra_transfer)
std_transfer = np.std(data_1000_intra_transfer)
best_fit_transfer = max([(norm, 'norm'), (expon, 'expon'), (gamma, 'gamma'), (lognorm, 'lognorm')],
                        key=lambda x: x[0].fit(data_1000_intra_transfer, loc=mean_transfer, scale=std_transfer))[1]

kde_no_transfer = sm.nonparametric.KDEUnivariate(data_1000_notransfer)
kde_no_transfer.fit()
mean_no_transfer = np.mean(data_1000_notransfer)
std_no_transfer = np.std(data_1000_notransfer)
best_fit_no_transfer = max([(norm, 'norm'), (expon, 'expon'), (gamma, 'gamma'), (lognorm, 'lognorm')],
                           key=lambda x: x[0].fit(data_1000_notransfer, loc=mean_no_transfer, scale=std_no_transfer))[1]

# Print the best fit distribution names
print(f"Transfer Data Best Fit Distribution: {best_fit_transfer}")
print(f"No Transfer Data Best Fit Distribution: {best_fit_no_transfer}")

# Adjust the spacing between subplots
plt.tight_layout()

# Save the figure
plt.savefig('distribution_kde_plot.jpeg', dpi=300)

# Display the figure
plt.show()
```

## when N= 1000

```python
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy.stats import norm, expon, gamma, lognorm
import numpy as np

# Read data from 'data_1000_Notransfer.txt'
data_1000_notransfer = []
with open('data_1000_Intra_Notransfer.txt', 'r') as file:
    for line in file:
        entry = line.strip().split(',')
        cn, time = int(entry[0]), float(entry[1])
        if time >= 64:
            data_1000_notransfer.append(cn)

# Read data from 'data_1000_Intra_transfer.txt'
data_1000_intra_transfer = []
with open('data_1000_Intra_transfer.txt', 'r') as file:
    for line in file:
        entry = line.strip().split(',')
        cn, time = int(entry[0]), float(entry[1])
        if time >= 64:
            data_1000_intra_transfer.append(cn)

# Set up the figure with four subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot histogram and KDE for transfer data - First Entry
axes[0, 0].hist(data_1000_intra_transfer, bins=10, color='blue', edgecolor='black',alpha=0.7, density=True)
axes[0, 0].set_xlabel('Number of Cryptococcus Cells')
axes[0, 0].set_ylabel('Density')
axes[0, 0].set_xscale('log')
axes[0, 0].set_title('Transfer (Histrogram)')

sns.kdeplot(ax=axes[0, 1], data=data_1000_intra_transfer, color='blue', fill=True)
axes[0, 1].set_xlabel('Number of Cryptococcus Cells')
axes[0, 1].set_ylabel('Density')
axes[0, 1].set_xscale('log')
axes[0, 1].set_title('Transfer (Density Plot)')

# Plot histogram and KDE for transfer data - No Transfer
axes[1, 0].hist(data_1000_notransfer, bins=10, color='green', edgecolor='black',alpha=0.7, density=True)
axes[1, 0].set_xlabel('Number of Cryptococcus Cells')
axes[1, 0].set_ylabel('Density')
axes[1, 0].set_xscale('log')
axes[1, 0].set_title('No Transfer')

sns.kdeplot(ax=axes[1, 1], data=data_1000_notransfer, color='green', fill=True)
axes[1, 1].set_xlabel('Number of Cryptococcus Cells')
axes[1, 1].set_ylabel('Density')
axes[1, 1].set_title('No Transfer (Density Plot)')

# Adjust the spacing between subplots
plt.tight_layout()

# Save the figure
plt.savefig('subplot_distribution_plots.jpeg', dpi=300)

# Display the figure
plt.show()
```

### SIgnificance test

```python
from scipy.stats import mannwhitneyu

# Perform Mann-Whitney U test

statistic, p_value = mannwhitneyu(data_1000_notransfer, data_1000_intra_transfer)

# Print the results
print("Mann-Whitney U statistic:", statistic)
print("p-value:", p_value)
```

## Mean distribution analysis


```python
import statistics

# Read data from 'data_1000_Notransfer.txt' and compute the mean and standard deviation
data_1000_notransfer = []
with open('data_1000_Intra_Notransfer.txt', 'r') as file:
    for line in file:
        entry = line.strip().split(',')
        cn, time = int(entry[0]), float(entry[1])
        if time >= 64:
            data_1000_notransfer.append(cn)

mean_notransfer = statistics.mean(data_1000_notransfer)
std_dev_notransfer = statistics.stdev(data_1000_notransfer)

# Read data from 'data_1000_Intra_transfer.txt' and compute the mean and standard deviation
data_1000_intra_transfer = []
with open('data_1000_Intra_transfer.txt', 'r') as file:
    for line in file:
        entry = line.strip().split(',')
        cn, time = int(entry[0]), float(entry[1])
        if time >= 64:
            data_1000_intra_transfer.append(cn)

mean_intra_transfer = statistics.mean(data_1000_intra_transfer)
std_dev_intra_transfer = statistics.stdev(data_1000_intra_transfer)
print('For No Transfer')
print("Mean:", mean_notransfer)
print("Standard Deviation:", std_dev_notransfer)

print('For Transfer')
print("Mean:", mean_intra_transfer)
print("Standard Deviation:", std_dev_intra_transfer)
```

from the mean and standard devation of the data, we can deduce that at \(N_0 = 0\), we have a distinguishable difference in the transfer and without transfer simulation result
